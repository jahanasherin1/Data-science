```python
import matplotlib.pyplot as plt
import numpy as np

#1st subplot
plt.figure(figsize=(12, 6))
x=np.array([1,4])
y=np.array([2,7])

plt.subplot(2,3,1)
plt.plot(x,y,'o')
font1 = {'family':'serif','color':'blue','size':10}
plt.title("Title1")
plt.xlabel("x-label",fontdict = font1)
plt.ylabel("y-label",fontdict = font1)

#2nd subplot
y1=np.array([0, 3, 0, 3, 0])
y2=np.array([3, 0, 2, 0, 3]) 

plt.subplot(2,3,2)
plt.plot(y1,marker='*',lw=2, color='orange')
plt.plot(y2 ,ls='--',lw=2, color='b')

#3rd subplot
x=np.array([1,3,7,8])
y=np.array([2,7,3,9])

plt.subplot(2,3,3)
plt.plot(x,y,marker='o', mec='black' ,mfc='yellow')

#4th subplot
x=np.array([1,4])
y=np.array([2,7])

plt.subplot(2,3,4)
plt.plot(x,y)
plt.grid()
plt.title("Title2")

#5th subplot
x = np.random.randint(20, size=(20))
y = np.random.randint(20, size=(20))
colors = np.random.randint(20, size=(20))
sizes = 10 * np.random.randint(20, size=(20))

plt.subplot(2,3,5)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,cmap='nipy_spectral')

#6th subplot
x=np.array(["A","B","C"])
y=np.array([2,6,9])
plt.subplot(2,3,6)

plt.bar(x,y,width = 0.1,color='r')

plt.subplots_adjust(
    left=None, right=None,bottom=None,top=None,
    wspace=0.8,
    hspace=0.6
)

plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 20, 20, 25])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]

plt.pie(y, labels = mylabels,explode = myexplode, startangle = 90,shadow=True)
plt.legend()
plt.show() 
```
